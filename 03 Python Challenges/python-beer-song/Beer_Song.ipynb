## Beer Song

Write a program which produces the lyrics to that beloved classic, that
field-trip favorite: 99 Bottles of Beer on the Wall.

Note that not all verses are identical.

See Readme for lyrics

### Algorithm

```python
def beer()->str:
    bot = {i: (str(i) + ' bottles') for i in range(0,100)} # dict with i : i bottles, from 0 to 99
    bot[0] = 'no more bottles' # 0 bottles => no more bottles
    bot[1] = '1 bottle' # 1 bottles => 1 bottle
    bot = dict(sorted(bot.items(), reverse=True)) # reverse so keys from 99 to 0
    for i in range(0,100):
        j = 99 - i
        if j == 0: # for last verse
            nex_bot = bot[99] # start again fom 99
            do = 'Go to the store and buy some more, '
        else: # for all verses but last
            nex_bot = bot[j-1] # count down
            do = 'Take one down and pass it around, '
        verse = bot[j].capitalize() + ' of beer on the wall, ' + bot[j] + ' bottles of beer.\n' + do +  nex_bot + ' of beer on the wall.\n'
        print(verse)

beer()
```
